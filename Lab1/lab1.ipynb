```python
address_book = {}


def print_all_contacts():
    for name, details in address_book.items():
        print(f"Name: {name}, Phone: {details['last_name']}, Email: {details['email']}")


def add_contact(name, last_name, email):
    address_book[name] = {'last_name': last_name, 'email': email}


def search_contact(name):
    if name in address_book:
        print(f"Name: {name}, Phone: {address_book[name]['last_name']}, Email: {address_book[name]['email']}")
    else:
        print(f"Contact {name} not found.")


add_contact('Jan', 'Kowalski', 'alice@example.com')
add_contact('Andrzej', 'Nowak', 'bob@example.com')
print_all_contacts()
search_contact('Andrzej')
search_contact('Michal')
```

```python
class Position:
    def __init__(self, position_x=0, position_y=0):
        self.position_x = position_x
        self.position_y = position_y

    def print_position(self):
        print(f"Position: ({self.position_x}, {self.position_y})")


class Rocket(Position):
    def __init__(self, name, fuel, position_x=0, position_y=0):
        super().__init__(position_x, position_y)
        self.name = name
        self.fuel = fuel

    def move(self, delta_x, delta_y):
        if self.fuel > 0:
            self.position_x += delta_x
            self.position_y += delta_y
            self.fuel -= 1
            if self.fuel < 0:
                self.explode()
        else:
            self.explode()

    def print_status(self):
        print(f"Rocket {self.name} - Fuel: {self.fuel}")
        self.print_position()

    def explode(self):
        print(f"Rocket {self.name} has exploded due to lack of fuel!")


# Initialize a collection of Rockets
rockets = [
    Rocket(name="Apollo", fuel=5),
    Rocket(name="Falcon", fuel=3),
    Rocket(name="Dragon", fuel=7)
]

# Iterate over the collection and move the rockets
for rocket in rockets:
    while rocket.fuel > 0:
        rocket.move(1, 1)
        rocket.print_status()
```

```python
x = np.linspace(0, 2.0, 100)
y = np.power(x, np.power(x, x))

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x) = x^(x^x)')
plt.title('Plot of f(x) = x^(x^x) for x in [0, 2.0]')
plt.grid(True)
plt.show()
```

```python
from scipy.stats import beta, skew, kurtosis

rng = np.random.default_rng(42)
# Generate the beta distribution
beta_dist = rng.beta(5, 2, size=100000)

# Calculate the mean
mean = beta_dist.mean()

# Calculate the standard deviation
std_dev = beta_dist.std()

# Calculate the skewness
skewness = skew(beta_dist)

# Calculate the kurtosis
kurt = kurtosis(beta_dist)

print(f"Mean: {mean}")
print(f"Standard Deviation: {std_dev}")
print(f"Skewness: {skewness}")
print(f"Kurtosis: {kurt}")

# Determine the skewness direction
if skewness > 0:
    skew_direction = "right skewed"
elif skewness < 0:
    skew_direction = "left skewed"
else:
    skew_direction = "symmetrical"

print(f"The distribution is {skew_direction}.")

# Interpret the kurtosis value
if kurt > 0:
    kurtosis_interpretation = "leptokurtic (more peaked than normal distribution)"
elif kurt < 0:
    kurtosis_interpretation = "platykurtic (less peaked than normal distribution)"
else:
    kurtosis_interpretation = "mesokurtic (similar to normal distribution)"

print(f"The kurtosis value indicates that the distribution is {kurtosis_interpretation}.")
```

```python
def rastrigin(X, Y):
    return 10 * 2 + (X**2 - 10 * np.cos(2 * np.pi * X)) + (Y**2 - 10 * np.cos(2 * np.pi * Y))

x = np.linspace(-5.12, 5.12, 400)
y = np.linspace(-5.12, 5.12, 400)
X, Y = np.meshgrid(x, y)
Z = rastrigin(X, Y)

plt.imshow(Z, extent=[-5.12, 5.12, -5.12, 5.12], origin='lower', cmap='viridis')
plt.colorbar()
plt.title('Rastrigin Function')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
